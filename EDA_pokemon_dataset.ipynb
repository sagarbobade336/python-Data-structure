7)What is the average value for each stat (e.g., hp, attack, defense, etc.)?

8)Which PokÃ©mon has the highest and lowest attack?

9)Is there a correlation between attack and weight?



```python
def get_stat(x, stat):
    try:
        return int([i.split('=')[1] for i in x.split(', ') if i.startswith(stat)][0])
    except:
        return None
```

```python
#7
df["stats"].apply(lambda x: get_stat(x, "hp"))
```

```python
df["stats"].apply(lambda x: get_stat(x, "attack"))
```

```python
df["stats"].apply(lambda x: get_stat(x, "defense"))
```

```python
#7
df["attack"] = df["stats"].apply(lambda x: get_stat(x, "attack"))
```

```python
df.loc[df["attack"].idxmax(), ["name", "attack"]]
```

```python
df.loc[df["attack"].idxmin(), ["name", "attack"]]
```

```python
#9
df["attack"].corr(df["weight"])
```
