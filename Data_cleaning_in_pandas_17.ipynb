```python
import pandas as pd
```

```python
df=pd.read_excel('/content/Customer Call List.xlsx')
```

```python
df
```

```python
#drop_duplicate = delect duplicate value
#inplace=True = it will make change in data dirctly without override
df.drop_duplicates(inplace=True)
df
```

```python
# removing columns if want multiple columns delete give list
df.drop(columns='Not_Useful_Column',inplace=True)
```

```python
df
```

```python
#strip = it well remove unwanted symbole
df["Last_Name"]= df['Last_Name'].str.strip("/..._")
df
```

```python
df["Phone_Number"]= df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
```

```python
df["Paying Customer"]=df['Paying Customer'].str.replace('Yes','Y')
df["Paying Customer"]=df['Paying Customer'].str.replace('No','N')
df
```
