Scatter Plot

```python
#Defining the data
roll_no = [1,2,3,4,5,6,7,8,9]
marks = [47,65,96,45,75,74,98,58,10]
```

```python
plt.style.use("dark_background")
```

```python
# Implementing Basic Scatter Plot
plt.figure(figsize=(12,6)) #..... change size of figure
plt.scatter(roll_no,marks)
plt.show
```

```python
# changing Color of the plotted Points
plt.scatter(roll_no,marks,color="orange")
plt.show
```

```python
# changing Marker of the Plotted Points
plt.scatter(roll_no,marks,color="red",marker="v")
plt.show()
```

```python
#More fuctions using Plot Function
plt.figure(figsize=(12,8))
plt.plot(roll_no,marks,"gv",markersize=20)
plt.show
```

```python
# Multiple Plot on same Figure
temperature_pune = [25,34,21,45,28,6,43,18,7,2]
humidity_pune = [28,25,29,20,26,50,19,29,52,55]

temperature_bangalore = [34,35,36,37,28,27,26,25,31,20]
humidity_bangalore = [40,38,36,35,42,44,14,40,34,45]
```

```python
plt.figure(figsize=(12,8))
plt.xticks(np.arange(0,60,5))
plt.yticks(np.arange(10,60,5))
plt.plot(temperature_pune,humidity_pune,"y*",markersize=15)
plt.plot(temperature_bangalore,humidity_bangalore,"gv",markersize=10)
plt.xlabel("Tempreture")
plt.ylabel("Humidity")
plt.show()
```
