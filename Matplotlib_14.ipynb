```python
import matplotlib.pyplot as plt
Name=["Sagar","Tejshree","Varun","Mayuri","Harshada","Nilesh"]
Marks=[25,36,45,64,58,64]

plt.plot(Name,Marks,color="b",label="Maths Mark",linestyle="-",marker="*")

plt.xlabel("Name")
plt.ylabel("Marks")
plt.title("Student Marks")
plt.legend()
plt.grid(alpha = 0.1)

plt.show()

```
