```python
#What is the average price of all cars in the dataset?
Sum = sum(df["Price"])
Avg = Sum/10000
Avg
```

```python
#Which car brand has the highest average price?
highest_avg_brand = df.groupby('Brand')['Price'].mean().idxmax()
highest_avg_brand
```

```python
#What is the most common fuel type in the dataset?
most_common_fuel_type = df['Fuel_Type'].value_counts().idxmax()
most_common_fuel_type
```

```python
#How many cars have a mileage greater than 200,000?
df[df["Mileage"]>200000]
```

```python
#What is the distribution of cars by transmission type?
df["Transmission"].value_counts()
```

```python
#Which car model has the highest price?
Max = max(df["Price"])
df[df["Price"]==Max]
```

```python
#What is the average price of cars by fuel type?
s=df.groupby("Fuel_Type")
s["Price"].mean()
```

```python
#How many cars are from the year 2020 or later?
df[df["Year"]>=2020]
```

```python
#What is the average engine size for each brand?
s=df.groupby("Brand")
s["Engine_Size"].mean()
```

```python
#Which car has the lowest mileage?
Min = min(df["Mileage"])
df[df["Mileage"]==Min]
```

```python
#What is the correlation between mileage and price?
df["Mileage"].corr(df["Price"])
```

```python
#How many cars have more than 2 owners?
df[df["Owner_Count"]>2]
```

```python
#What is the average price of cars by year?
df.groupby('Year')['Price'].mean()
```

```python
#Which car has the highest number of doors?
df[df["Doors"]==max(df["Doors"])]
```

```python
#What is the average mileage for each fuel type?
df.groupby("Fuel_Type")["Mileage"].mean()
```

```python
#How many cars are from the brand 'Toyota'?
len(df[df["Brand"]=="Toyota"])
```

```python
#What is the average price of cars with automatic transmission?
df[df["Transmission"]=="Automatic"]["Price"].mean()
```

```python
#Which car has the highest owner count?
df[df["Owner_Count"]==max(df["Owner_Count"])]
```

```python
#What is the average engine size for cars with diesel fuel type?
df[df['Fuel_Type'] == 'Diesel']['Engine_Size'].mean()
```

```python
#How many cars have a price greater than $10,000?
df[df["Price"]>=10000]
```
